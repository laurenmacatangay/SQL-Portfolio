```python
import bq_helper 

# The data set name needs to be passed to the Big Query Helper in order for us to 
# interact with the dataset. The variable was set to the same name as the dataset so we can 
# refer to it in our SQL code like we normally would if not pulling from Big Query
hacker_news = bq_helper.BigQueryHelper(active_project= "bigquery-public-data", 
                                       dataset_name = "hacker_news")
# This is where we actually begin using the data. "query" is the name we selected for the query.
# This will select the column "score" from the table "full" which is in the dataset "hacker_news"
query = """SELECT score
            FROM `bigquery-public-data.hacker_news.full`
            WHERE type = "job" """
# Keeping queries stored as variables allows you to do further analysis without rewriting
# the full code
job_post_scores = hacker_news.query_to_pandas_safe(query)
job_post_scores.score.mean()
```
